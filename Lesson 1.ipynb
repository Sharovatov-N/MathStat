#### 1. Из колоды в 52 карты извлекаются случайным образом 4 карты. a) Найти вероятность того, что все карты – крести. б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

Это задача комбинаторики на сочетание (порядок не важен):$$C_n^m=\frac{n!}{(n-m)!m!}$$Находим общее число равновозможных исходов  (полная группа событий):$$n=C_{52}^4=\frac{49*50*51*52}{2*3*4}=49*25*17*13=270725$$
a) Наступленине события А.<br>
Находим число благоприятсвующих исходов:$$m=T_{13}^4=\frac{10*11*12*13}{2*3*4}=5*11*13=715$$
Находим вероятность наступления события А:$$P(A)=\frac{m}{n}=\frac{715}{270725}=0.0026$$
б) Наступленине события B (один туз ИЛИ два туза ИЛИ три туза ИЛИ четыре туза).<br>
Находим число благоприятсвующих исходов:<br>
$$m=m_1+m_2+m_3+m_4$$
* $$m_1=T_4^1*C_{48}^3$$ $$T_4^1=4$$ $$C_{48}^3=\frac{46*47*48}{2*3}=23*47*16=17296$$ $m_1=4*17296=69184$
* $$m_2=T_4^2*C_{48}^2$$ $$T_4^2=\frac{3*4}{2}=6$$ $$C_{48}^2=\frac{47*48}{2}=47*24=1128$$ $m_2=6*1128=6768$
* $$m_3=T_4^3*C_{48}^1$$ $$T_4^3=4$$ $$C_{48}^1=48$$ $m_3=4*48=192$
* $m_4=1$<br>
$$m=69184+6768+192+1=76145$$<br>
Находим вероятность наступления события B:$$P(B)=\frac{m}{n}=\frac{76145}{270725}=0.281$$

#### 2. На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

Это задача комбинаторики на сочетание (порядок не важен):$$C_n^m=\frac{n!}{(n-m)!m!}$$Находим общее число равновозможных исходов  (полная группа событий):$$n=C_{10}^3=\frac{8*9*10}{2*3}=4*3*10=120$$
Число благоприятсвующих исходов - $m=1$<br>
Находим вероятность наступления события:$$P(A)=\frac{m}{n}=\frac{1}{120}=0.008$$

#### 3. В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

Это задача комбинаторики на сочетание (порядок не важен):$$C_n^m=\frac{n!}{(n-m)!m!}$$Находим общее число равновозможных исходов  (полная группа событий):$$n=C_{15}^3=\frac{13*14*15}{2*3}=13*7*5=455$$
Находим число благоприятсвующих исходов:$$m=C_{9}^3=\frac{7*8*9}{2*3}=7*4*3=84$$
Находим вероятность наступления события А:$$P(A)=\frac{m}{n}=\frac{84}{455}=0.1846$$

#### 4. В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

Это задача комбинаторики на сочетание (порядок не важен):$$C_n^m=\frac{n!}{(n-m)!m!}$$Находим общее число равновозможных исходов  (полная группа событий):$$n=C_{100}^2=\frac{99*100}{2}=99*50=4950$$
Число благоприятсвующих исходов - $m=1$<br>
Находим вероятность наступления события:$$P(A)=\frac{m}{n}=\frac{1}{4950}=0.0002$$
