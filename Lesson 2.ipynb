### Урок 2.

#### 1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

Из условия задачи это - биномиальное распределение, поэтому ипользуем формулу Бернулли: $$P_n(X=k) = C_n^k p^k q^{n-k}$$  $$P_{100}(k=85) = C_{100}^{85}p^{85}q^{15} = \frac{100!}{85!(100-85)!} \cdot 0.8^{85} \cdot 0.2^{15} = 0.0481$$

#### 2. Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?

Из условия задачи это - распределение Пуассона: $$P_m \approx \frac{\lambda^m}{m!}e^{-\lambda}$$ Находим $\lambda$: $$\lambda = np = 5000 \cdot 0.0004 = 2$$ 
Вероятность для $m=0$ $$P_0  \approx \frac{2^0}{0!}e^{-2} \approx  0.1353$$
Вероятность для $m=2$ $$P_2  \approx \frac{2^2}{2!}e^{-2} \approx  0.2707$$

#### 3. Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?

Из условия задачи это - биномиальное распределение, поэтому ипользуем формулу Бернулли: $$P_n(X=k) = C_n^k p^k q^{n-k}$$  
$$P_{144}(k=70) = C_{144}^{70}p^{70}q^{74} = \frac{144!}{70!(100-70)!} \cdot 0.5^{144} = 0.0628$$

#### 4. В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что все мячи белые? Какова вероятность того, что ровно два мяча белые? Какова вероятность того, что хотя бы один мяч белый?

* Ящик №1<br>
Находим общее число равновозможных исходов  (полная группа событий):$$n=C_{10}^2=\frac{9\cdot10}{2}=45$$
Событие А - 2 белых + 0<br>
Находим число благоприятсвующих исходов:$$m=C_{7}^2=\frac{6\cdot7}{2}=21$$
Событие В - 1 белый +1<br>
Находим число благоприятсвующих исходов:$$m=C_{7}^1\cdot C_3^1=7\cdot3=21$$
Событие С - 0 белых + 2<br>
Находим число благоприятсвующих исходов:$$m=C_{3}^2=3$$
Находим вероятность наступления события А:$$P_1(A)=\frac{m}{n}=\frac{21}{45}=0.4667$$
Находим вероятность наступления события B:$$P_1(B)=\frac{m}{n}=\frac{21}{45}=0.4667$$
Находим вероятность наступления события C:$$P_1(C)=\frac{m}{n}=\frac{3}{45}=0.0667$$
* Ящик №2<br>
Находим общее число равновозможных исходов  (полная группа событий):$$n=C_{11}^2=\frac{10\cdot11}{2}=55$$
Событие А - 2 белых + 0<br>
Находим число благоприятсвующих исходов:$$m=C_{9}^2=\frac{8\cdot9}{2}=36$$
Событие В - 1 белый +1<br>
Находим число благоприятсвующих исходов:$$m=C_{9}^1\cdot C_2^1=9\cdot2=18$$
Событие С - 0 белых + 2<br>
Находим число благоприятсвующих исходов:$$m=C_{2}^2=1$$
Находим вероятность наступления события А:$$P_2(A)=\frac{m}{n}=\frac{36}{55}=0.6545$$
Находим вероятность наступления события B:$$P_2(B)=\frac{m}{n}=\frac{18}{55}=0.3273$$
Находим вероятность наступления события C:$$P_2(C)=\frac{m}{n}=\frac{1}{55}=0.0182$$
**Находим вероятность того, что все мячи белые:** $$P(D)=P_1(A)\cdot P_2(A)=0.3055$$
**Находим вероятность того, что ровно два мяча белые:** $$P(F)=P_1(A)\cdot P_2(C)+P_1(C)\cdot P_2(A)=0.2049$$
**Находим вероятность того, что хотя бы один мяч белый:** $$P(G)=P_1(B)\cdot P_2(C)+P_1(C)\cdot P_2(B)+P(D)+P_1(A)\cdot P_2(B)+P_1(B)\cdot P_2(A)+P(F)=0.9989$$
